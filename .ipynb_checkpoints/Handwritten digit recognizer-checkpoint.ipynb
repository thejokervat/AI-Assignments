```python
import pandas as pd
data_train=pd.read_csv("mnist_train.csv", header=None)
data_train
```

```python
from sklearn.model_selection import train_test_split
d_train,d_val = train_test_split(data_train,shuffle = True) 
len(d_train)
```

```python
d_train_label=d_train[0].astype('category')
d_train_label=pd.get_dummies(d_train_label)
d_val_label=d_val[0].astype('category')
d_val_label=pd.get_dummies(d_val_label)
d_trainset=d_train.drop(d_train.columns[[0]], axis=1)

d_valset=d_val.drop(d_val.columns[[0]], axis=1)
```

```python

print(d_trainset.head())
print(d_valset.head())
print(d_train_label.head())
print(d_val_label.head())
```

```python
from keras.models import Sequential
from keras.layers import Dense, Activation
from matplotlib import pyplot as plt
```

```python
def print_results(model, features, labels):
    results = model.evaluate(features, labels, batch_size=128)
    print("Loss: " + str(results[0]) + "\tAccuracy: " + str(results[1]))
```

```python
model1 = Sequential([
    Dense(80, input_shape=(784,)),
    Activation('relu'),
    Dense(40),Activation('relu'),
    Dense(20),Activation('relu'),
    Dense(10),
    Activation('softmax'),
])

model1.compile(optimizer='adadelta',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model1.summary()

history1 = model1.fit(d_trainset, d_train_label,epochs=20,validation_data=(d_valset,d_val_label), batch_size=128)

print("Training data")
print_results(model1, d_trainset, d_train_label)
print("Validation data")
print_results(model1, d_valset, d_val_label)
```

```python
model2 = Sequential([
    Dense(400, input_shape=(784,)),
    Activation('tanh'),
    Dense(10),
    Activation('softmax'),
])

model2.compile(optimizer='adadelta',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model2.summary()

history2= model2.fit(d_trainset, d_train_label,epochs=20,validation_data=(d_valset,d_val_label), batch_size=128)
print("Training data")
print_results(model2, d_trainset, d_train_label)
print("Validation data")
print_results(model2, d_valset, d_val_label)
```

```python
model3 = Sequential([
    Dense(392, input_shape=(784,)),
    Activation('relu'),
    Dense(196),Activation('relu'),
    Dense(98),Activation('relu'),
     Dense(49),Activation('relu'),
    Dense(20),Activation('relu'),
    Dense(10),
    Activation('softmax'),
])
model3.compile(optimizer='adadelta',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

model3.summary()

history3 = model3.fit(d_trainset, d_train_label,epochs=20,validation_data=(d_valset,d_val_label), batch_size=128)

print("Training data")
print_results(model3, d_trainset, d_train_label)
print("Validation data")
print_results(model3, d_valset, d_val_label)
```

```python
print("Best model : ")
model3.summary()

print("################################")
print("Training data")
print_results(model3, d_trainset, d_train_label)
print("Validation data")
print_results(model3, d_valset, d_val_label)

print("################################")
print("Graph: ")
plt.plot(history3.history['acc'])
plt.plot(history3.history['val_acc'])
plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

plt.plot(history3.history['loss'])
plt.plot(history3.history['val_loss'])
plt.title('model loss')
plt.ylabel('loss')
plt.xlabel('epoch')
plt.legend(['train', 'validation'], loc='upper left')
plt.show()

print("#################################")
print("Test data results: ")
test_data = pd.read_csv("mnist_test.csv", header=None)
d_test_label=test_data[0].astype('category')
d_test_label=pd.get_dummies(d_test_label)
d_testset=test_data.drop(test_data.columns[[0]], axis=1)
print_results(model3,d_testset,d_test_label)
```
